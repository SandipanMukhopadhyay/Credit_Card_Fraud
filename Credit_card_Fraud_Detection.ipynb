Importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loadng the dataset into pandas dataframe
credit_card_data = pd.read_csv('/content/creditcard.csv')
```

```python
#first five rows of the dataset
credit_card_data.head()
```

```python
#last five rows of the dataset
credit_card_data.tail()
```

```python
# dataset information
credit_card_data.info()
```

```python
# checking the number of missing values in each column
credit_card_data.isnull().sum()
```

```python
# checking the distribution of legit transaction and fradulent transaction
credit_card_data['Class'].value_counts()
```

This dataset is highly unbalanced

0 --> Normal Transaction
1 --> Fradulent Transaction

```python
# separating the data for analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# statistical measures of the data
legit.Amount.describe()
```

```python
# statistical measures of the data
fraud.Amount.describe()
```

```python
# Compare the values for both transaction
credit_card_data.groupby('Class').mean
```

Under-sampling

Build a sample dataset containing similar distribution of normal transactions and Fradulent Transactions

Number of fradulent transaction -->492

```python
legit_sample = legit.sample(n=492)
```

Concatinating Two DataFrames

```python
new_dataset = pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Splitting the data into Features and Target

```python
X = new_dataset.drop(columns='Class',axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into Training data and Testing data

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Model Training

```python
model = LogisticRegression()
```

```python
# training the Logistic Regression Model with training data
model.fit(X_train,Y_train)
```

Model Evaluation

Accuracy Score

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print('Accuracy score on training data: ',training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print('Accuracy score on test data: ',test_data_accuracy)
```
